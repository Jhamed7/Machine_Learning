```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 50

man_hair = np.random.uniform(0, 30, (n_samples,1))  # hair length in cm
woman_hair = np.random.uniform(20, 60, (n_samples,1))

man_label = np.zeros((n_samples,1))
woman_label = np.ones((n_samples,1))

X = np.vstack((man_hair, woman_hair))
y = np.vstack((man_label, woman_label))


plt.scatter(X,y,s=None, marker='o',color='g')

plt.show()
```

```python
plt.scatter(X,y,s=None, marker='o',color='g')

for M in np.random.uniform(-0.05,0.06,(10,)):
    y_pred_1 = X*M
    plt.plot(X,y_pred_1,color='r')

plt.ylim(-0.5,1.5)
plt.show()
```

### My Adaline

```python
from numpy.linalg import inv

x = X.reshape(100, 1)
y = y.reshape(100, 1)
m_ada = np.matmul(inv(np.matmul(x.T, x)), np.matmul(x.T, y))
```

```python
m_ada
```

```python
plt.scatter(X,y,s=None, marker='o',color='g')

y_pred_1 = X*m_ada
plt.plot(X,y_pred_1,color='r', lw=3)

plt.show()
```
