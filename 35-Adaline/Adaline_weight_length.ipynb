```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200
x = np.arange(0,n,1,dtype=np.float64)

m = np.random.uniform(0.3,0.5,(n,))
b = np.random.uniform(5,10,(n,))

y = x*m

plt.scatter(x,y,s=None, marker='o',color='g')

plt.show()
```

```python
plt.scatter(x,y,s=None, marker='o',color='g')

for M in np.random.uniform(-0.2,0.8,(10,)):
    y_pred_1 = x*M
    plt.plot(x,y_pred_1,color='r')

plt.show()
```

### My Adaline

```python
# Adaline
from numpy.linalg import inv

x = x.reshape(200, 1)
y = y.reshape(200, 1)
m_ada = np.matmul(inv(np.matmul(x.T, x)), np.matmul(x.T, y))
```

```python
m_ada
```

```python
plt.scatter(x,y,s=None, marker='o',color='g')

y_pred_1 = x*m_ada
plt.plot(x,y_pred_1,color='r', lw=3)

plt.show()
```

### Scipy's Adaline

```python
from scipy import stats

a = stats.linregress(x.reshape((-1,)), y.reshape((-1,)))
```
